```python
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow  # Colab 전용 imshow 함수
import os

# 1. YOLO 모델 로드
model_path = '/content/drive/MyDrive/AIFFEL/Aiffelthon/ModelTest/best_pt_January_3rd_Jihye.pt'  # pt 모델 경로
model = YOLO(model_path)

# 2. 이미지 탐지 수행
image_path = '/content/drive/MyDrive/AIFFEL/Aiffelthon/ModelTest/onlyball1.png'  # 입력 이미지 경로
results = model.predict(source=image_path, save=True)  # 탐지 수행 및 결과 저장

# 3. 탐지된 이미지 저장 경로 확인
output_dir = str(results[0].save_dir)  # 저장된 결과 디렉토리 경로
print(f"탐지된 이미지가 저장된 디렉토리: {output_dir}")

# 4. 탐지된 이미지가 저장된 디렉토리의 파일 목록 확인
print("현재 디렉토리에 저장된 파일 목록:")
print(os.listdir(output_dir))

# 5. 절대 경로 설정 (저장된 이미지 경로 설정)
detected_img_path = f"{output_dir}/onlyball1.png"  # 탐지된 이미지 파일 절대 경로
print(f"탐지된 이미지 경로: {detected_img_path}")

# 6. 탐지된 이미지 시각화
if os.path.exists(detected_img_path):  # 이미지 파일 존재 여부 확인
    detected_image = cv2.imread(detected_img_path)
    if detected_image is not None:
        print("탐지된 이미지 출력 중...")
        cv2_imshow(detected_image)  # Colab 전용 이미지 출력
    else:
        print("Error: 이미지를 불러오지 못했습니다.")
else:
    print(f"Error: {detected_img_path} 파일이 존재하지 않습니다.")
```
