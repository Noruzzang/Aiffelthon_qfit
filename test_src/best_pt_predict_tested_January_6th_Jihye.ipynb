```python
import os
from ultralytics import YOLO

# 학습된 모델 로드
model = YOLO('/content/drive/MyDrive/AIFFEL_2024/Q-FIT/aiffelthon_January_6th/learning_result/weights/best.pt')  # 최적 모델 로드

# 테스트할 이미지가 있는 폴더 경로
test_images_folder = '/content/drive/MyDrive/AIFFEL_2024/Q-FIT/aiffelthon_January_6th/data/data_2/test/images'

# 텍스트 파일을 저장할 폴더 경로
save_folder = '/content/drive/MyDrive/AIFFEL_2024/Q-FIT/aiffelthon_January_6th/output'

# save_folder 경로가 존재하지 않으면 생성
if not os.path.exists(save_folder):
    os.makedirs(save_folder)

# 이미지 파일 목록 가져오기
image_files = [f for f in os.listdir(test_images_folder) if f.endswith(('.jpg', '.jpeg', '.png'))]

# 각 이미지에 대해 예측 실행
for image_file in image_files:
    # 이미지 경로
    image_path = os.path.join(test_images_folder, image_file)

    # 예측 실행
    results = model(image_path)

    # 결과가 리스트로 반환되므로, 각 결과에 대해 show() 실행
    for result in results:
        # 예측 결과를 시각화해서 보여줌
        result.show()

        # 예측 결과를 이미지로 저장
        result.save()

        # 바운딩 박스를 텍스트로 저장
        boxes = result.boxes  # 바운딩 박스 정보 가져오기
        if boxes is not None:
            # 예측된 각 바운딩 박스를 텍스트로 저장 (class_id, x_center, y_center, width, height)
            output_txt_file = os.path.join(save_folder, f'{os.path.splitext(image_file)[0]}.txt')
            with open(output_txt_file, 'w') as f:
                for box in boxes:
                    # xywh 형식으로 바운딩 박스 좌표를 추출 (4개의 값)
                    x_center, y_center, width, height = box.xywh[0]
                    class_id = int(box.cls)  # 클래스 ID를 추출
                    f.write(f"{class_id} {x_center} {y_center} {width} {height}\n")
```
